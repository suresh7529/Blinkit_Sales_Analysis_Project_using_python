## Data Cleaning

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat',
                                                        'low fat': 'Low Fat',
                                                        'reg': 'Regular'})
```

## **BUSINESS REQUIREMENTS**

## **KPI's REQUIREMENTS**

## 1. TOTAL SALES

```python
#Total Sales
total_sales = df['Sales'].sum()

#Average Sales
avg_sales = df['Sales'].mean()

#No of Item Sold
no_of_item_sold = df['Sales'].count()

#Average Rating
avg_ratings = df['Rating'].mean()

#display
print(f"Total Sales: ${total_sales: ,.0f}")
print(f"Average Sales: ${avg_sales: ,.1f}")
print(f"No of Item Sold: ${no_of_item_sold: ,.0f}")
print(f"Average Rating: {avg_ratings: ,.1f}")
```

## **CHART REQUIREMENTS**

## 1. TOTAL SALES BY FAT CONTENT

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels = sales_by_fat.index, autopct = '%.0f%%',
       startangle = 90)

plt.title('Sales By Fat Content')
plt.axis('equal')
plt.show()
```

## TOTAL SALES BY ITEM TYPE

```python
df.head()
```

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending = False)

plt.figure(figsize = (12,8))
bars = plt.bar(sales_by_type.index, sales_by_type.values)
plt.xticks(rotation = 90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales By Item Type')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom',fontsize = 10)

plt.tight_layout()
plt.show()
```

## FAT CONTENT BY OUTLET FOR TOTAL SALES

```python
total_sales_by_outlet = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
total_sales_by_outlet = total_sales_by_outlet[['Regular', 'Low Fat']]

ax = total_sales_by_outlet.plot(kind = 'bar',figsize=(8,5), title = 'Outlet Tier By Item Fat Content')
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title = 'Item Fat Content')
plt.tight_layout()
plt.show()
```

## TOTAL SALES BY OUTLET ESTABLISHMENT

```python
df.head()
```

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
plt.figure(figsize = (10,6))
plt.plot(sales_by_year.index, sales_by_year.values, marker = 'o',linestyle = '-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment ')

for x,y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x,y,f'{y:,.0f}', ha = 'center', va = 'bottom', fontsize = 10)
plt.tight_layout()
plt.show()
    
```

## SALES BY OUTLET SIZE

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize = (4,4))
plt.pie(sales_by_size, labels = sales_by_size.index, autopct = '%1.1f%%', startangle = 90)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

## SALES BY OUTLET LOCATION

```python
df.head()
```

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending = False)

plt.figure(figsize = (8,3))
ax = sns.barplot(x= 'Sales', y = 'Outlet Location Type', data = sales_by_location)

plt.title('Total Sales By Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()
plt.show()
```

## TOTAL SALES BY OUTLET IDENTIFIER

```python
sales_by_year = df.groupby('Outlet Identifier')['Sales'].sum().sort_index()
plt.figure(figsize = (10,6))
plt.plot(sales_by_year.index, sales_by_year.values, marker = 'o',linestyle = '-')

plt.xlabel('Outlet Identifier')
plt.ylabel('Total Sales')
plt.title('Outlet Identifier ')

for x,y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x,y,f'{y:,.0f}', ha = 'center', va = 'bottom', fontsize = 10)
plt.tight_layout()
plt.show()
    
```

## TOTAL SALES BY ITEM WEIGHT

```python
df.head()
```

```python
sales_by_weight = df.groupby('Item Weight')['Sales'].sum().sort_values(ascending = False).head(4)

plt.figure(figsize = (10,6))
bars = plt.bar(sales_by_weight.index, sales_by_weight.values)
plt.xticks(rotation = 90)
plt.xlabel('Item Weight')
plt.ylabel('Total Sales')
plt.title('Total Sales By Item Weight')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom',fontsize = 10)

plt.tight_layout()
plt.show()
```

## TOTAL SALES BY EACH OUTLET TYPE 

```python
sales_by_size = df.groupby('Outlet Type')['Sales'].sum()

plt.figure(figsize = (4,4))
plt.pie(sales_by_size, labels = sales_by_size.index, autopct = '%1.1f%%', startangle = 90)
plt.title('Outlet Type')
plt.tight_layout()
plt.show()
```
